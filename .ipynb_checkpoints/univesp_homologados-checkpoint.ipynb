```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "http://vestibular.univesp.br/3nGhChKn3_Homologados/db/UNIVESP-homologados.json?_=1575523516238"
data = requests.get(url)
datastore = data.json()
univesp_homologados = pd.DataFrame(datastore["RECORDS"])
univesp_homologados.head()
```

```python
univesp_homologados.shape
```

```python
filter_1 = univesp_homologados['Eixo'] == 'EIXO DE COMPUTAÇÃO'
filter_2 = univesp_homologados['Polo_1'] == 'SÃO PAULO - BUTANTÃ'
filter_3 = univesp_homologados['Polo_1'] == 'SÃO PAULO - JARDIM PAULISTANO'

print('Candidato x vagas no BUTANTÃ -',
      str(round(univesp_homologados[filter_1 & filter_2].shape[0] / 35)))
print('Candidato x vagas no JARDIM PAULISTANO -',
      str(round(univesp_homologados[filter_1 & filter_3].shape[0] / 35)))
```

```python
# univesp_homologados[filter_1]['Polo_1'].value_counts()

import requests
from bs4 import BeautifulSoup as BS

url = 'https://univesp.br/noticias/inscricoes-para-o-vestibular-2020-estao-encerradas#.XelUY5NKgon'
html = requests.get(url).text
soup = BS(html, 'lxml')
data = []
for wrapper in soup.find_all('p', {"class":"MsoNormal"})[23:]:
    data.append(wrapper.text)
    
pd.DataFrame(data)
```

```python
concorrentes = univesp_homologados[filter_1
                                   & (filter_2 | filter_3)].sort_values(
                                       by=['Nome'])
print(concorrentes.shape)
print(round(concorrentes.shape[0] / 70))
```
